```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Step 1 - Create the data
# Make a empty dictionary that maps the column names year and total_births to empty lists:
# Loop over the years from 1880 to 2020
# Use your parse_dataset(year) function to parse each year
# Use the pandas.sum() method to calculate the sum of the frequencies (births)
# Append the result to the dictionary using births_dict['total_births'].append()
# Also append the year to births_dict['year']
# Create a yearly_births DataFrame by passing births_dict to pd.DataFrame()
# This should result in a DataFrame that has the year as one column and the total_births in that year in another column.
```

```python
births_dict = {'year':[], 'total_births':[]}
def parse_dataset(year=1880):
    #import pandas as pd
    df_birth_table = pd.read_csv(f'../../data/names/yob{year}.txt', header=None, names=['name', 'gender', 'frequency'])
    return df_birth_table
for year in range(1880, 2021):
    sum_year = parse_dataset(year)['frequency'].sum()
    births_dict['total_births'].append(sum_year)
    births_dict['year'].append(year)
yearly_births = pd.DataFrame(births_dict)
yearly_births
```

```python
# Step 2 - Plotting
# use pandas to make a quick plot of your results

plt.plot(yearly_births['year'], yearly_births['total_births']);
```

```python
# use matplotlib to make a more detailed plot by adding a legend, title and and axis labels

plt.figure(figsize=(20, 10))
plt.plot(yearly_births['year'], yearly_births['total_births'], label='total births per year', color='green', linewidth=5)
plt.legend()
plt.title('development of the birth rate', size=(20))
plt.xlabel('year', size=(20))
plt.ylabel('number of births', size=(20))
plt.ticklabel_format(style='plain') # showing the whole number on the axes OR divide the y-column to get mio
plt.grid(True);
```
