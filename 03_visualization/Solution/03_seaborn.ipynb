```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step 1
#Open empty jupyter notebook and read in the dataframe that contains your gapminder data

df = pd.read_csv('../data/gapminder_total.csv')
df.head()
```

```python
#Step 2

#Create subset of data for one year

df_subset = df[df['year']== 2000]
df_subset.head()
```

```python
#Step 4

#Plot the life expectancy vs fertility rate

sns.scatterplot(x='life_expectancy', y='fertility', data=df_subset)
plt.title(f'Life Expectancy vs Fertility Rate for the Year 2000')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate')
plt.show()

# Which will return a rudimentary plot:
```

```python
#Step 5

#Using matplotlib add a title and labels. Increase the plot size to make it more readable.

plt.figure(figsize=(10, 8))
sns.scatterplot(x='life_expectancy', y='fertility', hue='continent',data=df_subset)
plt.title('Life Expectancy vs Fertility Rate')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate');
```

```python
#Step 6

#Weight the size of the scatter points to the population of each country. This can be done using the size parameter in the sns.scatterplot() function.

#Tip: the code above is correct but the magnitude of the population is astronomically larger than that of the life expectancy and fertility rate. Use some math to fix this.

plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', 
                hue='continent',
                size='population',
                sizes=(50, 1000),
                data=df_subset)
plt.title('Life Expectancy vs Fertility Rate')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate');
```

```python
#Step 7

#Setting the axes to display the origin also gives us a more complete picture of what is going on.
plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', 
                hue='continent',
                size='population',
                sizes=(50, 1000),
                data=df_subset)
plt.title('Life Expectancy vs Fertility Rate')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate')
plt.axis([0, 85, 0, 9.5]);

```

```python
#Step 8

#Save the plot
plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', 
                hue='continent',
                size='population',
                sizes=(50, 1000),
                alpha= 0.7, # Dots transparency
                data=df_subset)
plt.title('Life Expectancy vs Fertility Rate')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate')
plt.axis([0, 85, 0, 9.5]);

plt.savefig('./expectancy_vs_fertility_plot.png')
```

```python
#BONUS:

#Try out different plotting themes and save the plots
plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', 
                hue='continent',
                size='population',
                sizes=(50, 1000),
                alpha= 0.7, # Dots transparency
                palette='mako',
                data=df_subset)
plt.title('Life Expectancy vs Fertility Rate')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility Rate')
plt.axis([0, 85, 0, 9.5]);

plt.savefig('./expectancy_vs_fertility_plot_mako.png')
```

```python
# Bar plot
plt.figure(figsize=(14,6))
sns.barplot(x='continent', y='life_expectancy', errorbar=None,
                palette='magma', order=['Africa','Asia','Australia and Oceania','Europe','North America','South America'],
                data=df_subset)
plt.xlabel('Continent')
plt.ylabel('Life expectancy')
plt.title("Continents' life expectancy in 1995")
plt.axis([-1, 6, 0, 80]);
plt.savefig('./plot_continent_life_exp_ordered.png')
```
