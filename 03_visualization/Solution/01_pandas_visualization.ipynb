```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('../../data/winequality-red.csv', sep=';')
df
```

```python
# 1. plot avg amount of chlorides for each 'quality' value 
# check how many/which unique value are in quality
print('quality values: ',df.quality.unique())
print('Total unique values: ',len(df.quality.unique()))

# Calculate average chlorides for each 'quality' value
unique_qualities= df['quality'].unique()
avg_chlorides = {}
for quality in unique_qualities:
    avg_chlorides[quality] = df[df['quality'] == quality]['chlorides'].mean()

#plot 
plt.bar(avg_chlorides.keys(), avg_chlorides.values(), color='orange')
plt.title('Average Amount of Chlorides by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Chlorides')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
 # 2. Plot the `alcohol` values against `pH` values (in a scatter plot). Does there appear to be any relationship between the two?

plt.scatter(df['alcohol'], df['pH'], marker='o', color = 'purple' ,alpha=0.7)
plt.title('Alcohol vs pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()
```

```python
#3. Plot `total_acidity` values against `pH` values. Does there appear to be any relationship between the two?
df['total_acidity']= df['fixed acidity']+ df['volatile acidity']+ df['citric acid']

plt.scatter(df['total_acidity'], df['pH'], marker='o', color = 'c' ,alpha=0.7)
plt.title('Total Acidity vs pH')
plt.xlabel('Total Acidity')
plt.ylabel('pH')
plt.show()
```

```python
#4. Plot a histogram of the `density` values. Are they evenly distributed within the data set?

plt.hist(df['density'])
plt.title('Density Distribution')
plt.show()
```

```python
#5. Plot a boxplot to look at the distribution of `citric acid`. 

plt.boxplot(df['citric acid'])
plt.title("Citric Acid Distribution")
plt.show()
```
